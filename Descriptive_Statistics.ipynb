```
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings("ignore")

iris = sns.load_dataset("iris")
X = iris.drop('species',axis=1)
x=X['sepal_length']
```

```
mean=x.sum()/len(x)
print('Mean: {}'.format(mean))
```

```
import statistics
print('Mean: {}'.format(x.mean()))
```

```
median = x.median()
print('Median: {}'.format(median))
```

```
mode=x.mode()

print('Mode: {}'.format(mode))

print('Mode value: ',mode[0])
```

```
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw= {"height_ratios": (0.2, 1)})
sns.boxplot(x, ax=ax_box)
ax_box.axvline(mean, color='r', linestyle='--')
ax_box.axvline(median, color='g', linestyle='-')
ax_box.axvline(mode[0], color='b', linestyle='-')
sns.distplot(x, ax=ax_hist)
ax_hist.axvline(mean, color='r', linestyle='--')
ax_hist.axvline(median, color='g', linestyle='-')
ax_hist.axvline(mode[0], color='b', linestyle='-')
plt.legend({'Mean':mean,'Median':median,'Mode':mode})
ax_box.set(xlabel='')
plt.show()
```

```
from scipy.stats import skew
print('Skewness: {}'.format(skew(x)))
```

```
r=x.max()-x.min()
print('Range: ',r)
```

```
fig1, ax1 = plt.subplots()
ax1.set_title('BoxPlot')
ax1.boxplot(x)
```

```
Q1=np.percentile(x, 25) 
Q2=np.percentile(x, 50)  
Q3=np.percentile(x, 75)
print('Q1: {}'.format(Q1))
print('Q2: {}'.format(Q2))
print('Q3: {}'.format(Q3))
```

```
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw= {"height_ratios": (0.2, 1)})
sns.boxplot(x, ax=ax_box)
ax_box.axvline(Q1, color='r', linestyle='--')
ax_box.axvline(Q2, color='g', linestyle='-')
ax_box.axvline(Q3, color='b', linestyle='-')
sns.distplot(x, ax=ax_hist)
ax_hist.axvline(Q1, color='r', linestyle='--')
ax_hist.axvline(Q2, color='g', linestyle='-')
ax_hist.axvline(Q3, color='b', linestyle='-')
plt.legend({'Q1':Q1,'Q2':Q2,'Q3':Q3})
ax_box.set(xlabel='')
plt.show()
```

```
IQR=Q3-Q1
print('Interquartile Range: ',IQR)
```

```
sigma=np.std(x)
print('Standard Deviation: ',sigma)
```
