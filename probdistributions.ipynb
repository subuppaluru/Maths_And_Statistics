#Binomial 

```
import math
def binomial_prob(n, p, x):
    b = (math.factorial(n)/(math.factorial(x)*math.factorial(n-x)))*(p**x)*((1-p)**(n-x))
    return b
# probability of at least 3 boys in 6 children
n = 6
p = 1.09/2.09
b = 0
for x in range(3,7):     # x=3,4,5,6
    b += binomial_prob(n, p, x)   
print("%.3f" %b)    # Answer = 0.696


# probability of at least 2 defects in a batch of 10
n = 10
p = 0.12
print(round(sum([binomial_prob(n, p, x) for x in range(2, 11)]), 3))   
```

#Geometric

```
def geometric_prob(p, x):
    g = (1-p)**(x-1) * p
    return g
# probability that defect is found on 5th batch
p = 1/3
x = 5
g = geometric_prob(p, x)  
print("%.3f" %g)    # Answer = 0.066

# probability that defect is found during the first 5 inspections
g = 0
for i in range(1,6):     # i=1,2,3,4,5
    g += geometric_prob(p, i)   
print("%.3f" %g)     # Answer = 0.868
```

#Poisson

```
from math import factorial, exp
# probability of 5 defects in an area
x = 5
miu = 2.5
poisson_prob = ((miu**x) * exp(-miu)) / factorial(x)
print("%.3f" %poisson_prob)   # Answer = 0.067
```

#Normal

```
import math
def normal_prob(miu, stdev, x):
    return 0.5 * (1 + math.erf((x-miu)/(stdev * 2**0.5)))
# probability of time less than 19.5 hrs
miu = 20 
stdev = 2
print( '%.3f' %normal_prob(miu, stdev, 19.5) )
# Answer = 0.401
# probability of time between 20 and 22 hrs
print( '%.3f' %(normal_prob(miu, stdev, 22) - normal_prob(miu, stdev, 20)) )
# Answer = 0.341
```

```
# To find P(S < weight_limit)
weight_limit = 9800
n = 49
miu = 205
miu_S = n*miu
stdev = 15
stdev_S = (n**0.5)*stdev
print( '%.4f' %normal_prob(miu_S, stdev_S, weight_limit) )
```
